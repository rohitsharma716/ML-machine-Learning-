```python
df =  pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("income in $")
plt.scatter(df.year,df.income)
```

```python
year =  df.drop('income',axis='columns')
year
```
